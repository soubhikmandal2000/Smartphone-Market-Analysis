Exploratory Data Analysis using Altair, Matplotlib, NumPy, Pandas & Seaborn libraries for Smartphone available in the Indian market with 08 attributes and 2647 samples taken from Flipkart and answered the following questions :

1. Types of Smartphone available in the market?
2. Number of Smartphone manufacturers in the market?
3. Types of Smartphone available in the market?
4. What are the top 10 Smartphone brands?
5. Which brand has the most Smartphone sets on the market?
6. Rating (Out of 5) of the Smartphone Brands available in the market?
7. Average selling price by Brand, Memory & Storage of the Smartphone?

**Importing Libraries and Datasets of Smartphone**

```python
import altair as alt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('C:/Users/SOUBHIK MANDAL/Desktop/jupyter notebook/Smartphone Market Analysis/Flipkart_mobile_brands_scraped_data.csv')
df
```

```python
df.dtypes
```

```python
df.nunique()
```

```python
df.describe().T
```

```python
colors = ["#0000FF","#00FF00","#FFFF00","#FF7F00","#FF0000","#4361EE"]
palette = sns.color_palette(palette = colors)

sns.palplot(palette, size =1)
plt.show()
```

**Brands of smartphone availables in the market with their Original Price vs Selling Price**

```python
df['Brand'].groupby(df['Brand']).count().sort_values(ascending=False).iloc[:10]
```

```python
alt.Chart(df).mark_point(filled=True,size=40).encode(
    x='Original Price',
    y='Selling Price',
    color='Brand',
    tooltip=['Brand', 'Selling Price', 'Original Price']
).interactive()
```

```python
df.groupby('Brand')['Selling Price'].mean().sort_values(ascending=False).apply(np.ceil)
```

```python
sns.barplot(x=df['Brand'], y=df['Selling Price'],palette=colors)
plt.xticks(rotation="vertical")
plt.show()
```

```python
df.groupby('Brand')['Original Price'].mean().sort_values(ascending=False).apply(np.ceil)
```

```python
sns.barplot(x=df['Brand'], y=df['Original Price'],palette=colors)
plt.xticks(rotation="vertical")
plt.show()
```

**Rating of the smartphone Brands availables in the market**

```python
round(df.groupby('Brand')['Rating'].mean().sort_values(ascending=False),1)
```

**Memory & Storages of the Television Brands availables in the market**

```python
df['Brand'].groupby(df['Memory']).count().sort_values(ascending=False)
```

```python
sns.barplot(x=df['Memory'], y=df['Selling Price'],palette=colors)
plt.xticks(rotation="vertical")
plt.show()
```

```python
alt.Chart(df).mark_point(filled=True,size=40).encode(
    x='Original Price',
    y='Selling Price',
    color='Memory',
    tooltip=['Memory', 'Selling Price', 'Original Price']
).interactive()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,7))
sns.countplot(y="Brand", hue="Memory", data=df, palette=colors,order=df["Brand"].value_counts().iloc[:5].index)
```

```python
sns.barplot(x=df['Storage'], y=df['Selling Price'],palette=colors)
plt.xticks(rotation="vertical")
plt.show()
```

```python
alt.Chart(df).mark_point(filled=True,size=40).encode(
    x='Original Price',
    y='Selling Price',
    color='Storage',
    tooltip=['Storage', 'Selling Price', 'Original Price']
).interactive()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,7))
sns.countplot(y="Brand", hue="Storage", data=df, palette=colors,order=df["Brand"].value_counts().iloc[:5].index)
```
